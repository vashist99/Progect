```python
scaled_dataset = pd.DataFrame(scale_data, columns=data_toscale.columns)
```

```python
scaled_dataset['filename'] = data['filename']
scaled_dataset['genre'] = data['genre']
```

```python
scaled_dataset.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
def extract_song_name(filename):
    parts = filename.split('._')
    song_name_with_extension = parts[1].split('.')[0] if len(parts) > 1 else 'Unknown'
    return song_name_with_extension.rsplit('.', 1)[0]
```

```python
df = pd.DataFrame(scaled_dataset)
```

```python
df['Song'] = df['filename'].apply(extract_song_name)
```

```python
# //print all unique songs
df['Song'].unique()
```

```python
from sklearn.model_selection import GroupShuffleSplit

gss = GroupShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
train_idx, test_idx = next(gss.split(df, groups=df['Song']))

# Create train and test DataFrames
train_df = df.iloc[train_idx]
test_df = df.iloc[test_idx]

# Print results
print("Train DataFrame:")
print(train_df)
print("\nTest DataFrame:")
print(test_df)
```

```python
train_df.dtypes
```

```python
model.fit(train_df.drop(columns = ['genre', 'filename', 'Song']), train_df['genre'])
```

```python
train_df['pred'] = model.predict(train_df.drop(columns = ['genre', 'filename', 'Song']))
test_df['pred'] = model.predict(test_df.drop(columns = ['genre', 'filename', 'Song']))
```

```python
from sklearn.metrics import accuracy_score
accuracy_train = accuracy_score(train_df['genre'], train_df['pred'])
accuracy_test = accuracy_score(test_df['genre'], test_df['pred'])
print(accuracy_train)
print(accuracy_test)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train_df['genre'] = label_encoder.fit_transform(train_df['genre'])
test_df['genre'] = label_encoder.transform(test_df['genre'])
```

```python
X_train = train_df.drop(columns=['genre', 'filename', 'Song', 'pred'])
y_train = train_df['genre']
X_test = test_df.drop(columns=['genre', 'filename', 'Song', 'pred'])
y_test = test_df['genre']
```

```python
import tensorflow as tf
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),  # Correct input shape
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=1000, verbose=1)
print("done")
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test Accuracy: {test_acc*100:.2f}%")
```
